```python

def fill_null(data):
    for f in binary_features:
        data[f] = data[f].fillna(0)
    return data.dropna().drop_duplicates()


def drop_redundant(data, features):
    for feature in features:
        data = data.drop(feature, 1)
    return data


def binarize(data):
    data["guest_is_not_the_customer"] = np.where(data["guest_is_not_the_customer"] == True, 1, 0)
    data["is_user_logged_in"] = np.where(data["is_user_logged_in"] == True, 1, 0)
    data["is_first_booking"] = np.where(data["is_first_booking"] == True, 1, 0)
    return data


def process_categorical(data):
    for f in dummies:
        data = pd.get_dummies(data, prefix=f + '_', columns=[f])
    data = binarize(data)
    return data





def process_cancellation_policy(data):
    return



def process_input(data):
    data = drop_redundant(data, redundant_features)
    data = fill_null(data)
    for feature in positive_features:
        data = data[data[feature] > 0]
    for feature in non_negative_features:
        data = data[data[feature] >= 0]
    for feature in binary_features:
        data = data[data[feature].isin([0, 1])]
    for feature in ranged_features:
        data = data[data[feature].isin(ranged_features[feature])]
    return data


def create_new_features(data):
    special_requests = (list(data.filter(regex="request")))
    data["special_requests"] = data[special_requests].sum(axis=1)
    dates = ["checkin_date", "checkout_date", "booking_datetime", "cancellation_datetime" , "hotel_live_date"]
    for col in dates:
        data[col] = pd.to_datetime(data[col])
    data["booking_days"] = (data['checkout_date'] - data['checkin_date']).dt.days
    data["cost_per_night"] = data["original_selling_amount"] /( data["booking_days"] * data[ 'no_of_adults'])
    data = data.drop("original_selling_amount" , 1)
    data["weekday_num_nights"] = np.busday_count( data['checkin_date'].values.astype('datetime64[D]'),data['checkout_date'].values.astype('datetime64[D]'))
    data["weekend_num_nights"] = (data["booking_days"] - data["weekday_num_nights"])
    data["booked_ahead"] = (data['checkin_date'] - data["booking_datetime"]).dt.days
    data["booked_ahead"].replace({-1: 0}, inplace=True)  # same-day bookings are valued 0.
    data = data.replace({"guest_nationality_country_name": ISO2})
    data["local stay"] = (data["guest_nationality_country_name"]  ==data["hotel_country_code"] )
    data["local stay"] = np.where(data["local stay"] == True, 1, 0)

    cols_to_drop = [ "booking_days","checkout_date", "checkin_date", "booking_datetime" ]
    data['cancellation_datetime'] = data['cancellation_datetime'].fillna(0)
    data["cancelled"] = np.where(data["cancellation_datetime"] != 0, 1, 0)
    data["no_of_children"] = np.where(data["no_of_children"] > 0, 1, 0)
    data["pay_now"] = np.where(data["charge_option"] == "Pay Now", 1, 0)
    data = data[ data["hotel_live_date"].dt.year> 2008]
    data["new_hotel"] = np.where(data["hotel_live_date"].dt.year> 2016, 1, 0)
    data["unknown_payment"] = np.where(data["original_payment_method"] == "UNKNOWN", 1, 0)
    data["Card"] = np.where(data["original_payment_type"] == "Credit Card", 1, 0)

    data = data.drop("original_payment_method", 1)
    data = data.drop("original_payment_type" , 1)
    data = data.drop("hotel_live_date", 1)

    data = data.drop("cancellation_datetime", 1)
    data = data.drop(columns=cols_to_drop)
    data = data.drop("charge_option" , 1)
    return data

def remove_outliers(data):
    data = data[data["booked_ahead"] < 200]
    data = data[data["booking_days"] < 20]
    data = data[data["no_of_adults"] < 5]
    data = data[data["no_of_extra_bed"] < 3]
    data = data[data["no_of_room"] < 6]
    data = data[  data["weekend_num_nights"] < 3]
    data = data[  data["weekday_num_nights"] < 6]
    data = data[  data["cost_per_night"] < 200]


    # data = data[data["original_selling_amount"] < 80000]
    return data

def group_sparce(df):
    for i in country:
        frequencies = df[i].value_counts(normalize=True, ascending=False)
        threshold = frequencies[(frequencies.cumsum() > 0.1).idxmax()]
        df[i] = df[i].mask(df[i].map(df[i].value_counts(normalize=True)) < 0.05, i + '_Other')

    for i in categorial:
        frequencies = df[i].value_counts(normalize=True, ascending=False)
        threshold = frequencies[(frequencies.cumsum() > 0.1).idxmax()]
        df[i] = df[i].mask(df[i].map(df[i].value_counts(normalize=True)) < 0.05, i + '_Other')


    return df



def load_data(filename: str):

    df = pd.read_csv(filename )
    df = create_new_features(df)
    df = process_input(df)
    df = group_sparce(df)
    df = process_categorical(df)
    df = remove_outliers(df)

    return df.drop("cancelled", 1), df.cancelled , df
```
